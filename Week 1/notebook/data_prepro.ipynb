```python
import pandas as pd
import numpy as np
import joblib
import os
import path
# print(pd.__version__)

CURRENT_DIR = os.getcwd()
csv_file_path = os.path.join(CURRENT_DIR, "../data/raw/EV_cars.csv")
data_file = pd.read_csv(csv_file_path)
```

```python
from sklearn.preprocessing import LabelEncoder

data_file.drop_duplicates(inplace=True)
data_file.dropna(inplace=True) 
data_file.drop("Car_name_link", axis=1, inplace=True)

encdr = LabelEncoder()
data_file['Car_name'] = encdr.fit_transform(data_file['Car_name'])
joblib.dump(encdr, "../src/encoder.pkl")

data_file.to_csv("../data/clean/EV_cars_clean.csv")
data_file.head(18)
```
